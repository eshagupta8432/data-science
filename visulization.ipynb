```python
data=[1,2,3,4,5,6,8]
fig, ax=plt.subplots(figsize=(4,4))
ax.set_xlabel('no.of elements', fontdict={'fontsize':26, 'color':'purple'})
ax.set_ylabel('y-value', fontdict={'fontsize':34, 'color':'blue'})
ax.grid(True)
ax.set_title("basics",loc="center")
plt.plot(data, marker="o", markersize=15, markeredgecolor="black", markerfacecolor="green")
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10, 1)  # x values
y = np.array([1, 2, 4, 3, 5, 7, 6, 8, 7, 9]) # y values


plt.step(x,y, where="pre",label='Stair-step plot', color='blue')


plt.xlabel('X-axis')

plt.title('Stair-Step Plot Example')


plt.grid(True)

plt.show()
```
